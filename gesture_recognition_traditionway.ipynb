```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pickle




from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR, SVC, LinearSVC, LinearSVR
from sklearn.metrics import accuracy_score
from tqdm import tqdm
```

```python
df = []
path = []
label = []
fns = [i for i in os.listdir('data/gesture') if 'from' in i]

for i in fns:
    path_tmp = ['data/gesture' + '/' + i + '/' + j for j in os.listdir('data/gesture' + '/' + i)]
    path.append(path_tmp)

def load_data(path):
    data = pd.read_csv(path, sep = ',', header=None)
    data.columns = ['time', 'sensor1', 'sensor2', 'sensor3', 'sensor4', 'label']
    print(path)
    data[['label']] = data[['label']].astype(int)
    return data

def extract_label(path):
    num = path.split("/")[2][0]
    return int(num)

for i in range(0, len(path)):
    for j in path[i]:
        df.append(load_data(j))
        label.append(extract_label(j))
```

# data prepare

```python
x=np.linspace(0, 99, 100)
# print(x)
def scaling(np_array):
    length = len(np_array)
    xp = np.linspace(0, length, length)
    xp = xp*100/length
    xp = xp.astype(int)

    y = np.interp(x, xp, np_array)
    y = y.astype(int)
    return y

def pd_to_np(df0):
    sen1 = scaling(df0['sensor1'])
    sen2 = scaling(df0['sensor2'])
    sen3 = scaling(df0['sensor3'])
    sen4 = scaling(df0['sensor4'])
    sensor = np.array((sen1, sen2, sen3, sen4))

    return sensor
```

```python
dataset = []
for i in range(len(df)):
    tmp = pd_to_np(df[i])
#     tmp = np.reshape(tmp, 400)
    dataset.append(tmp)

```

```python
# extract max index:
new_dataset = []
for i in range(len(dataset)):
    tmp = np.argmax(dataset[i], axis=1)
    new_dataset.append(tmp)
```

```python
new_dataset = np.array(new_dataset)
label = np.array(label)
label = np.reshape(label, (len(df), 1))
```

```python
Data_label = np.hstack((new_dataset, label))
# Data_label = np.concatenate((new_dataset,label),axis=1)
```

```python
Data_label[0:100, :]
```

```python
np.random.shuffle(Data_label)
# print(Data_label[:, -1])
Data = Data_label[:, :-1]
label = Data_label[:, -1]
print(label)
```

# Model

```python
X_train, X_test, y_train, y_test = train_test_split(Data, label)
```

```python
ln = RandomForestClassifier()
ln.fit(X_train, y_train)
pred = ln.predict(X_test)
accuracy_score(y_test, pred)
```

```python
ln = ExtraTreesClassifier()
ln.fit(X_train, y_train)
pred = ln.predict(X_test)
accuracy_score(y_test, pred)
```

```python
ln = SVC()
ln.fit(X_train, y_train)
pred = ln.predict(X_test)
accuracy_score(y_test, pred)
```
